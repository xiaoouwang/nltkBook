```python
# suppose gutenberg has been imported
for fileid in gutenberg.fileids():
    # raw = no split into tokens = how many letters
     num_chars = len(gutenberg.raw(fileid))
     num_words = len(gutenberg.words(fileid))
     num_sents = len(gutenberg.sents(fileid))
     num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    # average word length, average sentence length, lexical diversity
     print("word length: ", round(num_chars/num_words), " sentence length", round(num_words/num_sents), " lexical diversity", round(num_words/num_vocab), fileid)

```

```python
# brown corpus, text by genre
# http://icame.uib.no/brown/bcm-los.html
from nltk.corpus import brown
brown.categories()
brown.words(categories = 'news')
brown.words(fileids=['cg22'])
brown.sents(categories=['news', 'editorial', 'reviews'])
```

```python
# convenient resource for studying systematic differences between genres, a kind of linguistic inquiry known as stylistics
# use of modal verbs in news text
news_text = brown.words(categories='news')
# lower is one of the standard operations
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
     print(m + ':', fdist[m], end=' ')
```

```python
# conditional frequency distributions.
# the most frequent modal in the news genre is will, while the most frequent modal in the romance genre is could. 
cfd = nltk.ConditionalFreqDist(
           (genre, word)
           for genre in brown.categories()
           for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

```python
def generate_model(cfdist, word, num=15):
    # generate a texte of 15 words
    # and reset word to be the most likely token in that context (using max()); next time through the loop, we use that word as our new context.
    for i in range(num):
        print(word, end=' ')
        word = cfdist[word].max()

text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
# we treat each word as a condition
# and for each one we effectively create a frequency distribution over the following words
cfd = nltk.ConditionalFreqDist(bigrams)
cfd['living']
generate_model(cfd, 'living')
# list(nltk.bigrams(text))
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab
    return sorted(unusual)
# The Words Corpus is the /usr/share/dict/words file from Unix, used by some spell checkers
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
# takes some time, compute the percentage of real words
def content_fraction(text):
     stopwords = nltk.corpus.stopwords.words('english')
     content = [w for w in text if w.lower() not in stopwords]
     return len(content) / len(text)

content_fraction(nltk.corpus.reuters.words())
```

```python
# name corpus
names = nltk.corpus.names
names.fileids()
['female.txt', 'male.txt']
male_names = names.words('male.txt')
female_names = names.words('female.txt')
[w for w in male_names if w in female_names]
# It is well known that names ending in the letter a are almost always female.
cfd = nltk.ConditionalFreqDist(
           (fileid, name[-1])
           for fileid in names.fileids()
           for name in names.words(fileid))
cfd.plot()
# most names ending with a, e or i are female; 
# names ending in h and l are equally likely to be male or female; names ending in k, o, r, s, and t are likely to be male.
```

```python
# pronuncing dictionary 
# http://en.wikipedia.org/wiki/Arpabet 
# CMU Pronouncing Dictionary for US English, which was designed for use by speech synthesizers.
entries = nltk.corpus.cmudict.entries()
len(entries)

for entry in entries[42371:42379]:
     print(entry)

# word of three letters starting with p and ending with t
for word, pron in entries: 
     if len(pron) == 3: 
         ph1, ph2, ph3 = pron
         if ph1 == 'P' and ph3 == 'T':
             print(word, ph2, end=' ')
```

```python
# words end with nicks rhyme dictionary
syllable = ['N', 'IH0', 'K', 'S']
[word for word, pron in entries if pron[-4:] == syllable]
```

```python
# words ending with n but pronunced as m
[w for w, pron in entries if pron[-1] == 'M' and w[-1] == 'n']
# first two letters of words not starting with n but pronunced as N
sorted(set(w[:2] for w, pron in entries if pron[0] == 'N' and w[0] != 'n'))
```
